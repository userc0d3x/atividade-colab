## Análise Exploratória de Dados sobre Hábitos de Saúde e Bem-Estar

**Objetivo do Projeto**

O objetivo é realizar uma análise exploratória de dados sobre hábitos de saúde e bem-estar coletados por meio de um questionário.

A pesquisa foi realizada em novembro de 2025.

Ao começar a análise dos dados, percebo que a pesquisa traz uma visão geral dos hábitos de saúde dos participantes. A primeira coisa que noto é que os dados foram coletados em novembro de 2025 e incluem informações importantes, como idade, sono, atividade física e alimentação. Isso já me dá uma boa base para entender o comportamento geral das pessoas no questionário.

```python
dados.hist()
```

```python
dados["Quantos dias na semana você pratica Exercicio?"].hist()
```

Ao ver o código que importa o CSV do Google Sheets, percebo que o dataset está bem estruturado. Isso facilita muito a análise, porque não vou perder tempo limpando muitos erros.
```python
dados["Quantas vezes na semana você estuda em casa?"].hist()
```

```python
dados["Quantas vezes na semana você dorme depois das 22h?"].value_counts()
```

Quando olho para os histogramas (como o de idade, sono, etc.), percebo padrões claros. Esses gráficos ajudam a enxergar como os hábitos se distribuem entre os participantes. Me chamou atenção que alguns campos deveriam ser numéricos, mas vieram como texto, e isso pode ter atrapalhado a geração dos gráficos no notebook.
```python
dados.describe()
```

```python
media_exercicios = dados["Quantos dias na semana você pratica Exercicio?"].mean()
mediana_exercicios = dados["Quantos dias na semana você pratica Exercicio?"].median()

print(media_exercicios)
print(mediana_exercicios)
```

```python
media_ler = dados["Quantas vezes na semana você estuda em casa?"].mean()
mediana_ler = dados["Quantas vezes na semana você estuda em casa?"].median()

print(media_ler)
print(mediana_ler)
```
